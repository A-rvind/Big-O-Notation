```python
import matplotlib.pyplot as plt
import numpy as np

# Input size
x = np.linspace(1, 10, 100)

# Complexity functions
y_constant = np.ones_like(x)         # O(1)
y_logarithmic = np.log2(x)           # O(log n)
y_linear = x                         # O(n)
y_nlogn = x * np.log2(x)             # O(n log n)
y_quadratic = x**2                   # O(n^2)
y_exponential = 2**x                 # O(2^n)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y_constant, label="O(1)", linewidth=2)
plt.plot(x, y_logarithmic, label="O(log n)", linewidth=2)
plt.plot(x, y_linear, label="O(n)", linewidth=2)
plt.plot(x, y_nlogn, label="O(n log n)", linewidth=2)
plt.plot(x, y_quadratic, label="O(n²)", linewidth=2)
plt.plot(x, y_exponential, label="O(2ⁿ)", linewidth=2)

# Labels and legend
plt.title("Big O Complexity Graph", fontsize=16)
plt.xlabel("Input Size (n)", fontsize=12)
plt.ylabel("Operations", fontsize=12)
plt.ylim(0, 100)
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

# Here’s a graph of common Big O complexities:

O(1): Constant, flat line.

O(log n): Slowly growing curve.

O(n): Linear growth, straight line.

O(n log n): Slightly faster than linear but slower than quadratic.

O(n²): Steep quadratic growth.

O(2ⁿ): Rapid exponential growth.

This shows how dramatically runtime increases as the input size (n) grows, especially for exponential and quadratic complexities.
